```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = gpd.read_file(r'data/A_Zipcodes2010wData.shp')
bdf = pd.read_csv(r'data/newbusinesses.csv', parse_dates=[9])
bdf['year'] = pd.DatetimeIndex(bdf.entityformdate).year
bdf['month'] = pd.DatetimeIndex(bdf.entityformdate).month
bdf['day'] = pd.DatetimeIndex(bdf.entityformdate).day
```

```python
merged = df.merge(bdf, left_on='ZIPCODE', right_on='principalzipcode')
```

```python
region = [-109.5, -101.5, 36.75, 41.25]

for year in range(2012,2022,1):
    for month in range(1,13,1):
        plt.close()
    #print(merged[merged.entityformdate == single_date.strftime("%Y-%m-%d")].year)
        fig, ax = plt.subplots(1, figsize=(20,20))
        ax.set_xlim(region[0], region[1])
        ax.set_ylim(region[2], region[3])
        ax.set_title(f'{year} {month}', fontsize=25)
        df.plot(facecolor='None', edgecolor='lightgray', ax=ax)
        subdf = merged[(merged.month==month)&(merged.year==year)].dissolve(by='ZIPCODE', aggfunc='count')
        if subdf.empty:
            print(f'saving empty {year}_{month}')
            plt.savefig(f'data/results/{year}_{month}.png')
            plt.clf()
        else:
            subdf.plot(column='NAME10', vmin=0, vmax=50, ax=ax, legend=True)
            print(f'saving results from {year}_{month}')
            plt.savefig(f'data/results/{year}_{month}.png')
            plt.clf()
    
```
