```python
import geopandas as gpd
import matplotlib.colors as colors
import matplotlib.pyplot as plt

from numpy import log10
```

```python
df = gpd.read_file(r'data/Block_Data_2020.shp')
df.to_crs(epsg=26954, inplace=True)
df['logpop'] = log10(df['TOTALPOP'])
```

```python
fig, ax = plt.subplots(1, figsize=(20,20))
df[df.logpop>0].plot(column='logpop', linewidth=0,  ax=ax, cmap='Reds', legend=True)
ax.set_xlabel('Easting (m)')
ax.set_ylabel('Northing (m)')
ax.set_title('2020 Colorado Log10 Population by Census Block', fontsize=30)
fig.savefig('Day_12.png')
```
