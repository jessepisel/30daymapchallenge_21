```python
import geopandas as gpd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df= gpd.read_file(r'data/HighwayMilepostsinColorado.geojson')
```

```python
df.head()
```

```python
df.plot(column=df.collector,figsize=(15,15), alpha=0.8, legend=True, legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.35)}, cmap='tab20', markersize=0.125)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Highway Mileposts in Colorado by Collector', fontsize=20)
plt.text(-106.25, 36.9, r'Data From: https://data.colorado.gov/Transportation/Highway-Mileposts-in-Colorado/trm9-dm4m', {'color': 'w', 'fontsize': 10})
plt.savefig('Day_1.png')
```
