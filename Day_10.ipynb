```python
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
```

```python
sat = rasterio.open("data/11081757a.jpg")

fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=1, nrows=4, figsize=(10, 12), sharex=True)
fig.suptitle("Decoded NOAA-18 Weather Satellite Imagery", fontsize=20)

show((sat, 3), cmap='Reds', ax=ax1)
show((sat, 2), cmap='Greens', ax=ax2)
show((sat, 1), cmap='Blues', ax=ax3)
show(sat, ax=ax4)
# Add titles
ax1.set_title("Red")
ax2.set_title("Green")
ax3.set_title("Blue")
ax4.set_title("False Color")

plt.tight_layout()
plt.savefig("Day_10.png")
```
