```python
import pygmt
import geopandas as gpd
```

```python
df = gpd.read_file(r'data/irrigation/Div4_Irrig_2015.shp')
df.to_crs("EPSG:4326", inplace=True)
df.drop(['SW_WDID1', 'SW_WDID2',
       'SW_WDID3', 'SW_WDID4', 'SW_WDID5', 'SW_WDID6', 'SW_WDID7', 'SW_WDID8',
       'SW_WDID9', 'GW_TYPE1', 'GW_ID1', 'GW_CLASS1', 'GW_TYPE2', 'GW_ID2',
       'GW_CLASS2', 'GW_TYPE3', 'GW_ID3', 'GW_CLASS3', 'GW_TYPE4', 'GW_ID4',
       'GW_CLASS4', 'GW_TYPE5', 'GW_ID5', 'GW_CLASS5', 'GW_TYPE6', 'GW_ID6',
       'GW_CLASS6', 'GW_TYPE7', 'GW_ID7', 'GW_CLASS7', 'GW_TYPE8', 'GW_ID8',
       'GW_CLASS8', 'GW_TYPE9', 'GW_ID9', 'GW_CLASS9', 'GW_TYPE10', 'GW_ID10',
       'GW_CLASS10', 'GW_TYPE11', 'GW_ID11', 'GW_CLASS11', 'GW_TYPE12',
       'GW_ID12', 'GW_CLASS12', 'GW_TYPE13', 'GW_ID13', 'GW_CLASS13',
       'GW_TYPE14', 'GW_ID14', 'GW_CLASS14', 'GW_TYPE15', 'GW_ID15',
       'GW_CLASS15', 'GW_TYPE16', 'GW_ID16', 'GW_CLASS16', 'GW_TYPE17',
       'GW_ID17', 'GW_CLASS17', 'GW_TYPE18', 'GW_ID18', 'GW_CLASS18',
       'GW_TYPE19', 'GW_ID19', 'GW_CLASS19', 'GW_TYPE20', 'GW_ID20',
       'GW_CLASS20', 'IRR', 'COMMENTS', 'FARM_UNIT'], axis=1, inplace=True)
```

```python
region = [-109.5, -106.25, 37.75, 39.25]
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M4i", frame=["a",'+t"Gunnison Basin Crops"'])
fig.grdimage("@srtm_relief_03s", shading=True, cmap='oleron')
fig.plot(data=df[df.CROP_TYPE=='GRASS_PASTURE'], color='#a6cee3', label='Pasture')
fig.plot(data=df[df.CROP_TYPE=='DRY_BEANS'], color='#1f78b4', label='Beans')
fig.plot(data=df[df.CROP_TYPE=='ORCHARD_WITH_COVER'], color='#fccde5', label='CoveredOrchard')
fig.plot(data=df[df.CROP_TYPE=='ORCHARD_WO_COVER'], color='#bc80bd', label='UncoveredOrchard')
fig.plot(data=df[df.CROP_TYPE=='ALFALFA'], color='#fb9a99', label="Alfalfa")
fig.plot(data=df[df.CROP_TYPE=='CORN_GRAIN'], color='#ffff99', label="CornGrain")
fig.plot(data=df[df.CROP_TYPE=='GRAPES'], color='#fdbf6f', label="Grapes")
fig.plot(data=df[df.CROP_TYPE=='SWEET_CORN'], color='#ff7f00', label="SweetCorn")
fig.plot(data=df[df.CROP_TYPE=='SMALL_GRAINS'], color='#cab2d6', label="SmallGrains")
fig.plot(data=df[df.CROP_TYPE=='BLUEGRASS'], color='#6a3d9a', label="Bluegrass")
fig.plot(data=df[df.CROP_TYPE=='WHEAT_SPRING'], color='#e31a1c', label="SpringWheat")
fig.plot(data=df[df.CROP_TYPE=='SORGHUM_GRAIN'], color='#8dd3c7', label="SorghumGrain")
fig.plot(data=df[df.CROP_TYPE=='VEGETABLES'], color='#bebada', label="Vegetables")
fig.plot(data=df[df.CROP_TYPE=='POTATOES'], color='#80b1d3', label="Potatoes")
fig.plot(data=df[df.CROP_TYPE=='BARLEY'], color='#fdb462', label="Barley")
fig.plot(data=df[df.CROP_TYPE=='WHEAT_FALL'], color='#d9d9d9', label="FallWheat")

fig.text(text=["https://cdss.colorado.gov/gis-data/division-4-gunnison"], 
         x=-108.8, y=37.8, font="5p,Helvetica,white")
fig.legend(position="JTR+o1c/0c+w7c/0.5c",)
```
