```python
import pygmt
import geopandas as gpd
```

```python
df = gpd.read_file(r'data/irrigation/Div4_Irrig_2015.shp')
df.to_crs("EPSG:4326", inplace=True)
```

```python
region = [-109.5, -106.25, 37.75, 39.25]
fig = pygmt.Figure()
fig.basemap(region=region, projection="M4i", frame=["a",'+t"Gunnison Basin Irrigation"'])
fig.grdimage("@srtm_relief_03s", shading=True, cmap=['#A9D6E5','#2A6F97', '#012A4A' ])
fig.plot(data=df[df.IRRIG_TYPE=='FLOOD'], color='#012A4A', label='Flood')
fig.plot(data=df[df.IRRIG_TYPE=='FURROW'], color='#01497C', label='Furrow')
fig.plot(data=df[df.IRRIG_TYPE=='DRIP'], color='#2A6F97', label='Drip')
fig.plot(data=df[df.IRRIG_TYPE=='SPRINKLER'], color='#468FAF', label='Sprinkler')
fig.plot(data=df[df.IRRIG_TYPE=='GATED_PIPE'], color='#89C2D9', label="GatedPipe")
fig.plot(data=df[df.IRRIG_TYPE=='UNKNOWN'], color='#A9D6E5', label='Unknown')
fig.legend(position="JTR+o1c/0c+w7c/0.5c",)
fig.text(text=["https://cdss.colorado.gov/gis-data/division-4-gunnison"], x=-108.8, y=37.8, font="5p")
```

```python
fig.show(width=1000)
```
